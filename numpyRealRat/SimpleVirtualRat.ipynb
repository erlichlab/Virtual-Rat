```python
RNNs = {}
solvers = {}
choices = {}
probabilities = {}
logical_accuracies = {}
real_accuracies = {}

learning_rate = 1e-4
hidden_dim = 5


# To find the best config
min_bias = 10
best_size= 0
best_init_params = None

repeat = 10
####### Note the key of the dictionaries!
for ratname, rat in rats.iteritems():
    print ratname
    for size in xrange(1,2):
        for i in xrange(repeat):
            print "Batch size is %d" % (size,)
            print "Training for %d / %d" % (i+1,repeat)
            RNN = SimpleRNN(hidden_dim = hidden_dim)
            RNNs[i] = RNN
            solver = RNNsolver(RNN, rat.trainX, rat.trainY,
                               update_rule='sgd',
                               optim_config={'learning_rate': learning_rate,
                           }, num_epochs = 100,
                               lr_decay = 1,
                               batch_size = size,
                               verbose = True)
            solvers[i] = solver
            solver.train()
            choice, probs = rat.predict(RNN)
            probabilities[i] = probs
            choices[i] = choice
            acc = np.mean(choice == rat.valTrueY)
            accReal = np.mean(choice == rat.valY)
            logical_accuracies[i] = acc
            real_accuracies[i] = accReal

            print "Logical validation accuracy: %s, Real sequence validation accuracy: %s" % (acc, accReal)

            ##############
            sample_probabilities(probs, ratname, sample = 50)

            ##############
            loss_history(solver, ratname)

            #############
            sample_correct_rate(rat, sample = 500)

            # Plot for normalization
            trial_window = 3

            real_p2a, real_a2p = realRatSwitchCost(rats,trial_window = trial_window)
            p2a, a2p, p2a2, a2p2 = meanPerformance(rats, trial_window = trial_window)
    
            bias_p2a = bias(real_p2a, p2a)
            bias_a2p = bias(real_a2p, a2p)
            bias_mean = np.mean([bias_p2a,bias_a2p])
            
            rp2a = corr(real_p2a, p2a)
            ra2p = corr(real_a2p, a2p)
            r_mean = np.mean([rp2a,ra2p])
            
            if bias_mean < min_bias:
                min_bias = bias_mean
                best_r = r_mean
                best_size = size
                best_init_params = RNN.initparams
            elif bias_mean == min_bias and r_mean > best_r:
                min_bias = bias_mean
                best_r = r_mean
                best_size = size
                best_init_params = RNN.initparams
                
            
            print "The sum of square bias between the model and real rat's data on pro to anti is %f" % (bias_p2a,)
            print "The sum of square bias between the model and real rat's data on anti to pro is %f" % (bias_a2p,)
            print "The mean of two sum of square bias is %f" % (bias_mean,)
         
            print "The correlation coefficient between the model and real rat's data on pro to anti is %f" % (rp2a,)
            print "The correlation coefficient between the model and real rat's data on anti to pro is %f" % (ra2p,)
            print "The mean correlation coefficients is %f" % (r_mean,)

            draw_3d(real_p2a, real_a2p, p2a, a2p, p2a2 = None, a2p2 = None, exclude_cpv = False, trial_window = 3)

    print "The minimum bias is %f" % (min_bias,)
    print "The best correlation coefficient is %f" % (best_r,)
    print "The best batch size is %d" % (best_size,)
```

```python
simrat = SimRat(RNN)
BLOCKS = 3 # The maximum length of blocks
MINIMUM_BLOCK_LENGTH = [1,9,17]
RANGE = 7
BLOCK_NUM = 100
REPEAT = 20 # Number of repetition times
p2a_drops = np.zeros((BLOCKS+2))
a2p_drops = np.zeros((BLOCKS+2))
meanPerformances = np.zeros((BLOCKS+2))
meanPerformanceExceptSwitchAnti = np.zeros((BLOCKS+2))
meanPerformanceExceptSwitchPro = np.zeros((BLOCKS+2))
meanPerformanceSwitchAnti = np.zeros((BLOCKS+2))
meanPerformanceSwitchPro = np.zeros((BLOCKS+2))
p2a_drops[0] = None
a2p_drops[0] = None
meanPerformances[0] = None
meanPerformanceExceptSwitchAnti[0] = None
meanPerformanceExceptSwitchPro[0] = None
meanPerformanceSwitchAnti[0] = None
meanPerformanceSwitchPro[0] = None

p2a_drops[BLOCKS+1] = None
a2p_drops[BLOCKS+1] = None
meanPerformances[BLOCKS+1] = None
meanPerformanceExceptSwitchAnti[BLOCKS+1] = None
meanPerformanceExceptSwitchPro[BLOCKS+1] = None
meanPerformanceSwitchAnti[BLOCKS+1] = None
meanPerformanceSwitchPro[BLOCKS+1] = None

for j in xrange(BLOCKS):
    block_length = MINIMUM_BLOCK_LENGTH[j]
    p2a_drop = np.zeros((REPEAT))
    a2p_drop = np.zeros((REPEAT))
    meanPerformance = np.zeros((REPEAT))
    PerformanceExceptSwitchAnti = np.zeros((REPEAT))
    PerformanceExceptSwitchPro = np.zeros((REPEAT))
    PerformanceSwitchAnti = np.zeros((REPEAT))
    PerformanceSwitchPro = np.zeros((REPEAT))
    for i in xrange(REPEAT):
        X, y = randomLengthBlocks(block_length, BLOCK_NUM ,Range = RANGE)
        choice, probs = simrat.predict(X,y)
        p2a_drop[i] = simrat.p2a_prob[3] - simrat.meanPerformanceExceptSwitchAnti
        a2p_drop[i] = simrat.a2p_prob[3] - simrat.meanPerformanceExceptSwitchPro
        meanPerformance[i] = simrat.hit_rate.mean()
        PerformanceExceptSwitchAnti[i] = simrat.meanPerformanceExceptSwitchAnti
        PerformanceExceptSwitchPro[i] = simrat.meanPerformanceExceptSwitchPro
        PerformanceSwitchAnti = simrat.p2a_prob[3]
        PerformanceSwitchPro = simrat.a2p_prob[3]
        
    p2a_drops[j+1] = p2a_drop.mean()
    a2p_drops[j+1] = a2p_drop.mean()
    meanPerformances[j+1] = meanPerformance.mean()
    meanPerformanceExceptSwitchAnti[j+1] = PerformanceExceptSwitchAnti.mean()
    meanPerformanceExceptSwitchPro[j+1] = PerformanceExceptSwitchPro.mean()
    meanPerformanceSwitchAnti[j+1] = PerformanceSwitchAnti.mean()
    meanPerformanceSwitchPro[j+1] = PerformanceSwitchPro.mean()

XRANGE = range(0,BLOCKS+2)
p2aplot, = plt.plot(XRANGE, p2a_drops, 'ro--')
a2pplot, = plt.plot(XRANGE, a2p_drops, 'bo--')
plt.legend([p2aplot, a2pplot],["pro to anti","anti to pro"])
xticks = ["","1-8",'9-16',"17-24",""]
plt.xticks(XRANGE, xticks)
plt.xlabel('Block length')
plt.ylabel('Switch cost')
plt.title('Switch cost asymmetry as a function of last block length')
plt.show()

meanplot, = plt.plot(XRANGE, meanPerformances, 'ko-')
meanAnti, = plt.plot(XRANGE, meanPerformanceExceptSwitchAnti, 'ro-')
meanPro, = plt.plot(XRANGE, meanPerformanceExceptSwitchPro, 'bo-')
switchAnti, = plt.plot(XRANGE, meanPerformanceSwitchAnti,'ro--')
switchPro, = plt.plot(XRANGE, meanPerformanceSwitchPro, 'bo--')
plt.legend([meanplot, meanAnti, meanPro, switchAnti, switchPro],
           ["Mean block performance","Mean anti block performance except swtich","Mean pro block performance except swtich",
           "Mean pro to anti switch", "Mean anti to pro switch"], loc='center left', bbox_to_anchor=(1, 0.7))
plt.xticks(XRANGE, xticks)
plt.xlabel('Block length')
plt.ylabel('Mean Performance')
plt.title('Mean performances as a function of last block length')
plt.show()
```
