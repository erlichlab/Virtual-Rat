```python
import numpy as npp
import minpy.numpy as np
from minpy.nn import layers
from minpy.nn.model import ModelBase
from nn.solver import Solver
import matplotlib.pyplot as plt
import minpy
minpy.set_global_policy("only_numpy")
import cPickle
from VanillaRNN import vanillaRNN
from SimRat import SimRat
from trialGenerator import trialGenerator
from dataProcessFunctions import *

%matplotlib inline
plt.rcParams['figure.figsize'] = (8.0, 6.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
N=20
generator = trialGenerator(N)
block = 20
```

```python
trainX,trainY = generator.getSequence(length = 10000, mode="alternative",p2a=0.5, block_size = block)
valX, valY = generator.getSequence(length = 5000, mode="alternative",p2a=0.5, block_size = block)
```

```python
RNN = vanillaRNN(N=N, hidden_dim=2, reg=0, p = 1, decay_rate = 1)
```

```python
learning_rate = 8e-3
solver = Solver(RNN,
                trainX, trainY,
                valX, valY,
                num_epochs=30,
                batch_length = 10,
                init_rule='xavier',
                update_rule='adam',
                optim_config={
                      'learning_rate': learning_rate,
                    },
                lr_decay = 1,
                verbose=True,
               )
solver.init()
for i in range(5):
    solver.change_settings(learning_rate=learning_rate/2)
    solver.train()
    
    trial_window = 3
    rat = SimRat(RNN,trial_window=trial_window)
    heatmap(rat.RNN.h.asnumpy(),start=0,end=300)
    testX, testY = generator.getSequence(length = 5000, mode = "test", block_size = block, std=5)
    probs = rat.predict(testX,testY)
    ratname = 'SimRat' 
    loss_history(solver, ratname) 
    sample_probabilities(probs, ratname, sample = 50) 
    sample_correct_rate(rat, sample = 1000) 
    p2a, a2p = rat.p2a_prob, rat.a2p_prob
    draw_3d(p2a, a2p, trial_window = trial_window)
```

```python
trial_window = 3
rat = SimRat(RNN,trial_window=trial_window)
heatmap(rat.RNN.h.asnumpy(),start=0,end=500)
testX, testY = generator.getSequence(length = 5000, mode = "test", block_size = block, std=5)
probs = rat.predict(testX,testY,noise=0)
ratname = 'SimRat' 
loss_history(solver, ratname) 
sample_probabilities(probs, ratname, sample = 50) 
sample_correct_rate(rat, sample = 1000) 
p2a, a2p = rat.p2a_prob, rat.a2p_prob
draw_3d(p2a, a2p, trial_window = trial_window)
```
